#         PREPARAÇÃO DOS DADOS 

```python
df.duplicated().sum()
```

```python
def faixa_etaria(idade):
    if idade < 18:
        return "Menor de idade"
    elif idade < 30:
        return "Jovem"
    elif idade < 60:
        return "Adulto"
    else:
        return "Senior"

df["Faixa_Etaria"] = df["Age"].apply(faixa_etaria)
```

```python
df["Faixa_Etaria"].value_counts()
```

```python
df["Faixa_Etaria"].isnull().sum()
```
