```python
import pandas as pd

df = pd.read_csv('/content/base_original.csv', sep=None, engine='python')

print("Primeiras 5 linhas do DataFrame após correção do separador:")
display(df.head())

print("\nInformações gerais do DataFrame (tipos de dados e valores não nulos) após correção do separador:")
df.info()
```

Agora que os dados foram carregados corretamente, vamos remover as linhas duplicadas para garantir a unicidade dos registros.

```python
# Remove as linhas duplicadas
initial_rows = len(df)
df.drop_duplicates(inplace=True)

print(f"Linhas duplicadas removidas: {initial_rows - len(df)}")
print(f"Total de linhas após remover duplicatas: {len(df)}")
```

Com base na verificação anterior, não há valores nulos no DataFrame. Podemos seguir para a remoção de espaços em branco das colunas de texto, que foi uma das suas instruções de limpeza.

```python
# Remove os espaços em branco das colunas
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].str.strip()

print("Espaços em branco removidos das colunas de texto.")
# Exibe as colunas já limpas
display(df.head())
```

A próxima etapa é padronizar o texto em todas as colunas de tipo `object` (string). Vamos converter o texto para a primeira letra maiúscula para uma melhor consistência.

Para padronizar corretamente o nome e o sobrenome (ou seja, a primeira letra de cada palavra em maiúscula) na coluna 'Name', vamos aplicar a função `str.title()`.

```python
# Padroniza a coluna 'Name' para que o 'nome' e o 'sobrenome' comecem com a letra maiúscula.
df['Name'] = df['Name'].str.title()

print("Coluna 'Name' padronizada.")
# Exibe as primeiras linhas para confirmar
display(df.head())
```

A coluna `Last_Login` deve ser convertida para o tipo de dado `datetime` para permitir análises baseadas em tempo. Vamos fazer essa conversão agora.

```python
# A coluna 'Last_Login' é convertida para o tipo de dado 'datetime' para permitir análises baseadas em tempo.
df['Last_Login'] = pd.to_datetime(df['Last_Login'])

print("Tipo de dado da coluna 'Last_Login' convertido para datetime.")

df.info()

display(df.head())
```

## 2. Padronização de Nomes de Colunas

Para facilitar a compreensão e o manuseio dos dados, as colunas foram renomeadas para português, seguindo a convenção `Snake Case`.

```python
df.rename(columns={
    'User_ID': 'ID',
    'Name': 'Nome',
    'Age': 'Idade',
    'Country': 'País',
    'Subscription_Type': 'Assinatura',
    'Watch_Time_Hours': 'Horas_Assistidas',
    'Favorite_Genre': 'Gênero_Favorito',
    'Last_Login': 'Ultimo_'
}, inplace=True)

print("Colunas renomeadas com sucesso.")

display(df.head())
df.info()
```
